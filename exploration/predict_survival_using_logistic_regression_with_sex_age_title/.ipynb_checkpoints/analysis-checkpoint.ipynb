The 'Age' column is the only one with missing values with almost 20%. Because this is such a large percentage, simply excluding the NaN values is unadvisable. I will now explore the column to determine a sensible way to fill in the missing values.

```python
_ = plt.hist(df['Age'], bins = 30)
_ = plt.title('Age')
_ = plt.ylabel('Counts')
```

The values in the age column seem to have tails at both ends. I chose to use the Median age to fill the NaN values since it is less sensitive to outliers compared to the mean.

```python
df = df.fillna({'Age': df.Age.median()})
df.info()
```

```python
df = df.astype({
    'Age': 'float64',
    'Name': 'object',
    'Sex': 'category',
    'Survived': 'int64',
})
```

```python
df.info()
```

# Feature Engineering

```python
simplify_title = {
    "Capt": "Officer",
    "Col": "Officer",
    "Major": "Officer",
    "Jonkheer": "Royalty",
    "Don": "Royalty",
    "Sir": "Royalty",
    "Dr": "Officer",
    "Rev": "Officer",
    "the Countess": "Royalty",
    "Dona": "Royalty",
    "Mme": "Mrs",
    "Mlle": "Miss",
    "Ms": "Mrs",
    "Mr": "Mr",
    "Mrs": "Mrs",
    "Miss": "Miss",
    "Master": "Master",
    "Lady": "Royalty"
}

df['Title'] = df['Name'].apply(
    lambda full_name: (
        simplify_title[
            full_name.split(',')[1].split('.')[0].strip()
        
        ]
    )
)

df['Title'] = df['Title'].astype('category')
df.head()
```

# Modeling

```python
df = df[['Survived', 'Sex', 'Age', 'Title']]
```

```python
# One hot encoding categorical variables

df = pd.get_dummies(df, columns =['Sex', 'Title'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('Survived', axis = 1).values, 
    df['Survived'].values, 
    test_size = 0.2
)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# <li> Majority Vote Classifier

```python
majority_vote = round(y_train.mean())
y_test_predictions = [majority_vote] * len(y_test)
accuracy = accuracy_score(y_true=y_test, y_pred=y_test_predictions)
print('The prediction accuracy with the majority vote classifier is {:.1f}%'.format(accuracy * 100))
```

# <li> Logistic Regression

```python
lr = LogisticRegression(solver = 'lbfgs')
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('The prediction accuracy with the untuned logistic regression model is {:.1f}%'.format(accuracy * 100))
```
